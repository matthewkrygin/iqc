```python
from qiskit import *
from math import *
import random
import numpy as np
```

```python
#def get
def state1qset(qc, vectors, coeffs, iQ):
    init_state = vectors[0]*coeffs[0] + vectors[1]*coeffs[1]
    qc.initialize(init_state, iQ)
    
def state2qset(a):
    qc = QuantumCircuit(2, 2)
    u, s, vh = np.linalg.svd(a, full_matrices=True)
    print(u)
    print(s)
    print(vh)
    res1 = state1qset(qc, u.transpose(), s, 0)
    res2 = state1qset(qc, vh, s, 1)
    return qc.copy()
```

```python
circuit = QuantumCircuit(2, 2)
```

```python
a = [[ 1/2, 1/2], 
     [1/2, 1/2]]  #from presentation
u, s, vh = np.linalg.svd(a, full_matrices=True)
print(u)
print(s)
print(vh)
```

```python
state1qset(circuit, u.transpose(), s, 0)
```

```python
state1qset(circuit, vh, s, 1)
```

```python
circuit.draw('mpl')
```

```python
from qiskit.tools.visualization import plot_histogram, plot_state_city
simulator = Aer.get_backend('statevector_simulator')

# Execute and get counts
result = execute(circuit, simulator).result()
statevector = result.get_statevector(circuit)
plot_state_city(statevector)
```
