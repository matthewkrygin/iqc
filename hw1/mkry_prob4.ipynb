$$ \frac{2}{3} \Ket{00} + \frac{1}{3} \Ket{01} - \frac {2} {3} \Ket{11}  =

\begin{bmatrix} 
\alpha_0\beta_0 \\
\alpha_0\beta_1 \\

\alpha_1\beta_1

$$
\begin{cases}
\alpha_0\beta_0 = 2/3  & (1)\\
\alpha_0\beta_0 = 1/3  & (2)\\

\end{cases}

\begin{matrix}

\end{matrix} 
$$

$$
